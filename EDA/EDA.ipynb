```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../heart.csv')
```

```python
df.head()
```

```python
# pairplot of everything
sns.pairplot(data=df[['Age','RestingBP','Cholesterol','MaxHR','Oldpeak']])
```

```python
# missing data
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
# data which are zeros
sns.heatmap(df.eq(0),yticklabels=False,cbar=False,cmap='viridis')
```

```python
# The propotion between the people how have hear disease and people who don't
sns.set_style('whitegrid')
sns.countplot(x='HeartDisease',data=df,palette='RdBu_r')
```

```python
# The propotion between the people how have hear disease and people who don't and whether they are male or female
sns.set_style('whitegrid')
sns.countplot(x='HeartDisease',hue='Sex',data=df,palette='RdBu_r')
```

```python
# The propotion between the people how have hear disease and people who don't and their chest pain types
sns.set_style('whitegrid')
sns.countplot(x='HeartDisease',hue='ChestPainType',data=df,palette='rainbow')
plt.savefig('chestpaintype.png')
```

```python
# distribution of the age
df['Age'].hist(bins=30,color='darkred',alpha=0.7)
plt.xlabel('age')
plt.savefig('age.png')
```

```python
# a count plot of the resting electrocardiogram
sns.countplot(x='RestingECG',data=df)
plt.savefig('restingecg.png')
```
